```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import bnoutils as bno
import numpy as np
%matplotlib inline
```

```python
line_data = bno.BNOData('../Data/002_LineMotion.csv')
line_data.compute_global_position()

fig = plt.figure(figsize = (10, 10))
ax  = fig.add_subplot(111, projection = '3d')
ax.scatter(line_data.global_position[:, 0],
           line_data.global_position[:, 1],
           line_data.global_position[:, 2],
           color = plt.cm.viridis(np.linspace(0, 1, line_data.global_position.shape[0])),
          zdir = 'z');
ax.set_xlim(-0.025, 0.05)
ax.set_ylim(-0.025, 0.05)
ax.set_zlim(-0.025, 0.05)
ax.view_init(45, 10);
```

```python
square_data = bno.BNOData('../Data/003_SquareMotion.csv')
square_data.compute_global_position()

fig = plt.figure(figsize = (10, 10))
ax  = fig.add_subplot(111, projection = '3d')
ax.scatter(square_data.global_position[:, 0],
           -square_data.global_position[:, 1],
           square_data.global_position[:, 2],
           color = plt.cm.viridis(np.linspace(0, 1, line_data.global_position.shape[0])),
          zdir = 'z', s = 2);
ax.set_xlim(-3, 6)
ax.set_ylim(-3, 3)
ax.set_zlim(-3, 3)
ax.view_init(90, 180);
```
